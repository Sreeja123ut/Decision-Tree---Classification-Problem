```python
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("04+-+decisiontreeAdultIncome.csv")
data.head()
```

```python
#Check Datatype

data.dtypes
```

```python
#Copy the data

df = data.copy()

```

```python
# get dummies for catagorical variable

df1=pd.get_dummies(df,drop_first = True)
df1.head()
```

```python
df1.shape
```

```python
#create X and Y

X = df1.iloc[:,:-1]
Y = df1.iloc[:,-1]
```

```python
#Split the model into training and testing

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test =  train_test_split(X,Y,test_size = 0.3,random_state=1234)
```

```python
#Import Decision tree classifer

from sklearn.tree import DecisionTreeClassifier
```

```python
# Max Depth - the max split of the node count
#min_samples_split - total number of samples ( rows) that need to be split
#min_samples_leaf -  minimal outcome the tree should have for each node
#max_leaf_nodes - maximum outcomes the tree can have for each mode
#-------------------------
#Splitter - random split/best split ( we have to chose the best split. It's nothing but the information gain)
#max_features = number of variables/features to be compared
#presort - wether to sort the data or not before the split
#-----------------------
#criterion ( Gini/Entropy) - Entropy of the data ( which calculates the impurity/randomness in the data). We should have lower entrophy values, Gini 

#min_impurity_decrease - How much impurity decrease is expected before we split


```

```python
#Let's train the data
dtc = DecisionTreeClassifier(random_state = 1234)
dtc.fit(X_train,Y_train)
Y_pred = dtc.predict(X_test)
```

```python
# Evaluate the model

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,Y_pred)
Accuracy = dtc.score(X_test,Y_test)
```

```python
print(cm)
print(Accuracy)
```
